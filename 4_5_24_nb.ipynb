```python
# imports
import pandas as pd
from scipy import stats
```

```python
# Load datafile
wt = pd.read_csv('4.5.24.CH12WT.csv')
```

```python
# I chose to ignore "N"s in the reads
wt['ref_nt'] = wt['ref_nt'].apply(lambda x: x.strip()[0])
wt = wt.drop(columns=['FN'])
wt = wt.drop(columns=['RN'])
wt.head()
```

```python
# This loop compares the reads to the reference sequence and 
# calculates the mutation rate for each position in the sequence

wt_mut_rates = []

for index, row in wt.iterrows():
    if row['ref_nt'] == 'A':
        correct = row['FA'] + row['RA']
        mut = row['FC'] + row['FG'] + row['FT'] + row['RC'] + row['RG'] + row['RT']
    elif row['ref_nt'] == 'C':
        correct = row['FC'] + row['RC']
        mut = row['FA'] + row['FG'] + row['FT'] + row['RA'] + row['RG'] + row['RT']
    elif row['ref_nt'] == 'G':
        correct = row['FG'] + row['RG']
        mut = row['FA'] + row['FC'] + row['FT'] + row['RA'] + row['RC'] + row['RT']
    elif row['ref_nt'] == 'T':
        correct = row['FT'] + row['RT']
        mut = row['FA'] + row['FG'] + row['FC'] + row['RA'] + row['RG'] + row['RC']
    
    mut_rate = mut/(correct + mut)
    wt_mut_rates.append(mut_rate)
```

```python
wt_mut_rates
```

```python
# Calculate overall mutation rate for sample
avg_wt_mut_rate = sum(wt_mut_rates) / len(wt_mut_rates)
avg_wt_mut_rate
```
